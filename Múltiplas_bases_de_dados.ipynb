```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

#Tratando os valores do dataset
Tirando valores duplicados ou com o mesmo significado

Substituindo espaço em branco por _

Preenchendo valores nulos

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace("APP","Aplicativo")
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('bandeira').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.groupby('estado').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(" ","_")
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas[vendas["estado"].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas['estado'].fillna("MS", inplace=True)
vendas.info()
```

```python
media_preco = vendas['Preço'].mean()
vendas['Preço'].fillna(media_preco, inplace=True)
vendas.info()
```

```python
vendas[["Preço","Preço_com_frete"]].describe()
```

```python
preco_errado = vendas.query("Preço > Preço_com_frete")
vendas_correto = vendas.query("Preço < Preço_com_frete")
```

```python
vendas_correto.describe()
```

```python
vendas_correto.head()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

###1º Departamentos mais vendidos

```python
agg_vendas_departamento = vendas_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
agg_vendas_departamento
```

###2º Média de preço com frete por Nome de Departamento

```python
round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg("mean").sort_values(ascending=False).reset_index(),2)
```

###3º Quantidade de vendas por Mês

```python
vendas_correto.groupby("Data").idcompra.nunique().sort_values(ascending=False).reset_index()
```

criando a coluna de mês no dataset

```python
vendas_correto["mes"] = vendas_correto["Data"].dt.month
vendas_correto.head()
```

```python
vendas_correto.groupby("mes").idcompra.nunique().sort_values(ascending=False).reset_index()
```

###4º Média de renda para cada tipo de canal de venda

```python
cliente = pd.read_excel('cliente_varejo.xlsx')
cliente.head()
```

```python
cliente[["idade","renda"]].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({"renda":"float"})
cliente.info()
```

Criando dataset da união das duas tabelas

```python
vendas_cliente = vendas_correto.merge(cliente, on="cliente_Log", how="left")
vendas_cliente.head()
```

Resposta item 4

```python
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

###5º Média de idade de clientes por bandeira

```python
agg_bandeira_idade = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_bandeira_idade
```

# Biblioteca de gráficos em python

matplotlib.pyplot

seaborn

```python
import matplotlib.pyplot as plt
```

## Graficos de Barra

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_bandeira_idade["bandeira"],agg_bandeira_idade["idade"], color="green")
plt.title("Média de idade de clientes por bandeira", loc="center", fontsize=18, fontweight='bold', color='black')
plt.xlabel("Bandeira")
plt.ylabel("Média de idade")
plt.show()
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idcanal_renda["idcanalvenda"],agg_idcanal_renda["renda"], color="green")
plt.title("Média de renda para cada tipo de canal de venda", loc="center", fontsize=18, fontweight='bold', color='black')
plt.xlabel("Canal de venda")
plt.ylabel("Média de renda")
plt.show()
```

```python
agg_dept_preco = round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_dept_preco
```

```python
fig = plt.figure(figsize=(15,5))
plt.bar(agg_dept_preco["Nome_Departamento"],agg_dept_preco["Preço_com_frete"], color="green")
plt.title("Média de preço com frete por Nome de Departamento", loc="center", fontsize=18, fontweight='bold', color='black')
plt.xlabel("Departamento")
plt.ylabel("Preço com frete")
plt.show()
```

## Grafico de Linha

```python
venda_por_data = vendas_correto.groupby("Data").idcompra.nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize=(15,5))
plt.title("Quantidade de vendas por mês", loc="center", fontsize=18, fontweight='bold', color='black')
plt.xlabel("Mês")
plt.ylabel("Quantidade de vendas")
x = venda_por_data["Data"]
y = venda_por_data["idcompra"]
plt.plot(x,y, color="green", linewidth=2)
plt.show()
```
